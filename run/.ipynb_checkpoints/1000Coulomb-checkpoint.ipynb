First, let's do some imports.

```python
# Numpy
import numpy as np
import pandas as pd
%matplotlib nbagg
import matplotlib.pyplot as plt
import scipy.stats

# Machine learning
import tensorflow as tf
import keras
import sklearn
import sklearn.preprocessing, sklearn.base, sklearn.utils, sklearn.metrics

# Various Python tricks and libraries
import requests
import time
import functools
import operator
import collections
from tqdm import tqdm, tqdm_notebook, trange
```

```python
def r2_score(ytrue, ypred): # sklearn.metrics.r2_score in tensorflow. 1 output only. 

    ytrue_mean = tf.reduce_mean(ytrue, name="ytrue_mean")
    r2_score = tf.subtract(1., tf.truediv(tf.reduce_mean((ytrue - ypred) ** 2), tf.reduce_mean((ytrue - ytrue_mean) ** 2)), name="r2_score")
    return r2_score

class TqdmProgBar(keras.callbacks.Callback):
    def __init__(self, n_epochs, field, fig, ax, label=''): # logs[field] is updated in tqdm bar and plotted in fig, ax with label
        self.n_epochs = n_epochs
        self.field = field
        self.fig = fig
        self.ax = ax
        self.label = label
        self.logs = []
    def on_train_begin(self, logs):
        self.pbar = tqdm_notebook(total=self.n_epochs, leave=False)
    def on_train_end(self, logs):
        self.pbar.close()
        self.ax.plot(range(len(self.logs)), self.logs, label=self.label)
        self.ax.set_ylim(bottom=-0.1)
        self.fig.canvas.draw()
        self.ax.legend()
        plt.legend()
    def on_epoch_end(self, epoch, logs):
        self.pbar.set_description("%.2f" %(logs[self.field]))
        self.pbar.update(1)
        self.logs.append(logs[self.field])
```

Good. Now onto the real deal.

Get the data.

```python
N_SAMPLES = 1000
MAX = 50
MIN = 3

df = pd.DataFrame()
df[['x1','x2','x3']] = pd.DataFrame(np.random.rand(N_SAMPLES, 3)) * (MAX - MIN) + MAX
df['e'] = np.linalg.norm(df[['x1','x2','x3']], axis=-1) ** -1.
df[['y1','y2','y3']] = pd.DataFrame(
    df[['x1','x2','x3']].values / np.linalg.norm(df[['x1','x2','x3']], axis=-1, keepdims=True)
)
```

Prepare the data.

```python
df[['x1_scaled', 'x2_scaled', 'x3_scaled', 'e_scaled', 'y1_scaled', 'y2_scaled', 'y3_scaled']] = pd.DataFrame(
    sklearn.preprocessing.scale(df.copy())
)
```

Decide on a model, and train.

# Keras E

```python
fig, ax = plt.subplots(1,1)
```

```python
N_EPOCHS = 5000
BATCH_SIZE = 1024

# GRAPH
G = keras.models.Sequential([
    keras.layers.Dense(units=24, activation='tanh', input_shape=(3,)),
    keras.layers.Dense(units=20, activation='tanh'),
    keras.layers.Dense(units=16, activation='tanh'),
    keras.layers.Dense(units=1, activation=None)
])
G.compile(loss = 'mse', 
          metrics = [r2_score], 
          optimizer = keras.optimizers.SGD(lr=0.01, momentum=0.91, nesterov=True))

# SESSION
G.fit(
    x = df[['x1_scaled', 'x2_scaled', 'x3_scaled']], 
    y = df['e_scaled'].reshape(-1,1),
    epochs = N_EPOCHS,
    batch_size = BATCH_SIZE,
    validation_split = 0.2,
    shuffle = True,
    verbose = 2,
    callbacks = [TqdmProgBar(n_epochs=N_EPOCHS, field='r2_score', fig=fig, ax=ax)]
)
```

Visualize results.

```python
yhat = G.predict(x = df[['x1_scaled', 'x2_scaled', 'x3_scaled']]).reshape(-1)
y = df['e_scaled'].reshape(-1)
plt.scatter(y, yhat)
```

# Tensorflow E

```python
N_EPOCHS = 5000
BATCH_SIZE = 512

# GRAPH
tf.reset_default_graph()
X = tf.placeholder(name="X", dtype=tf.float32, shape=(None, 3))
E = tf.placeholder(name="E", dtype=tf.float32, shape=(None, 1))

h1 = tf.layers.dense(X, units=24, activation=tf.tanh, name="h1")
h2 = tf.layers.dense(h1, units=20, activation=tf.tanh, name="h2")
h3 = tf.layers.dense(h2, units=16, activation=tf.tanh, name="h3")
Ehat = tf.layers.dense(h3, units=1, activation=None, name="Ehat")

loss = tf.reduce_mean(tf.square(Ehat - E), keepdims=False)
training_op = tf.train.MomentumOptimizer(learning_rate=0.01, momentum=0.91, use_nesterov=True).minimize(loss)

# SESSION
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in tqdm_notebook(range(N_EPOCHS)):
        index = np.random.choice(range(N_SAMPLES), BATCH_SIZE, replace=False)
        sess.run(training_op, 
                 feed_dict = {
                     X: df.loc[index, ['x1_scaled', 'x2_scaled', 'x3_scaled']],
                     E: df.loc[index, ['e_scaled']]
                 })
        if epoch % 100 == 1:
            _Ehat = sess.run(Ehat, 
                             feed_dict = {
                                 X: df[['x1_scaled', 'x2_scaled', 'x3_scaled']]
                             })
            print sklearn.metrics.r2_score(df[['e_scaled']].values.reshape(-1), _Ehat.reshape(-1))
```

# Tensorflow F Use grad

Get the data.

```python
N_SAMPLES = 1000
MAX = 50
MIN = 3

df = pd.DataFrame()
df[['x1','x2','x3']] = pd.DataFrame(np.random.rand(N_SAMPLES, 3)) * (MAX - MIN) + MAX
df['e'] = np.linalg.norm(df[['x1','x2','x3']], axis=-1) ** -1.
df[['f1','f2','f3']] = pd.DataFrame(
    df[['x1','x2','x3']].values / np.linalg.norm(df[['x1','x2','x3']], axis=-1, keepdims=True)
)
```

Prepare the data.

```python
df[['x1_scaled', 'x2_scaled', 'x3_scaled', 'e_scaled', 'f1_scaled', 'f2_scaled', 'f3_scaled']] = pd.DataFrame(
    sklearn.preprocessing.StandardScaler().fit_transform(df.copy())
)
```

Train.

```python
N_EPOCHS = 5000
BATCH_SIZE = 512

# GRAPH
tf.reset_default_graph()
X = tf.placeholder(name="X", dtype=tf.float32, shape=(None, 3))
F = tf.placeholder(name="F", dtype=tf.float32, shape=(None, 3))

h1 = tf.layers.dense(X, units=24, activation=tf.tanh, name="h1")
h2 = tf.layers.dense(h1, units=20, activation=tf.tanh, name="h2")
h3 = tf.layers.dense(h2, units=16, activation=tf.tanh, name="h3")
Ehat = tf.layers.dense(h3, units=1, activation=None, name="Ehat")
Fhat = tf.gradients(Ehat, [X], name="Fhat")[0]

loss = tf.reduce_mean(tf.square(Fhat - F), keepdims=False)
training_op = tf.train.MomentumOptimizer(learning_rate=0.01, momentum=0.91, use_nesterov=True).minimize(loss)

# SESSION
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in tqdm_notebook(range(N_EPOCHS)):
        index = np.random.choice(range(N_SAMPLES), BATCH_SIZE, replace=False)
        sess.run(training_op, 
                 feed_dict = {
                     X: df.loc[index, ['x1_scaled', 'x2_scaled', 'x3_scaled']],
                     F: df.loc[index, ['f1_scaled', 'f2_scaled', 'f3_scaled']]
                 })
        if epoch % 100 == 1:
            _Fhat = sess.run(Fhat, 
                             feed_dict = {
                                 X: df[['x1_scaled', 'x2_scaled', 'x3_scaled']]
                             })
            print sklearn.metrics.r2_score(df[['f1_scaled']].values.reshape(-1), _Fhat[:,0].reshape(-1))
```

```python
_Xscaled_ij = sklearn.preprocessing.scale(_X_ij.reshape(-1,3)).reshape(NI, NJ, 3)
_Fscaled_i = sklearn.preprocessing.scale(_F_i) * NJ
```

```python
N_EPOCHS = 50000
BATCH_SIZE = 512
r2s = []
fig, ax = plt.subplots(1,1)
t = tqdm_notebook(range(N_EPOCHS))

# GRAPH
tf.reset_default_graph()
X_ij = tf.placeholder(name="X_ij", dtype=tf.float32, shape=(None, None, 3))
F_i = tf.placeholder(name="F_i", dtype=tf.float32, shape=(None, 3))

h1 = tf.layers.dense(X_ij, units=16, activation=tf.tanh, name="h1")
h2 = tf.layers.dense(h1, units=20, activation=tf.tanh, name="h2")
h3 = tf.layers.dense(h2, units=32, activation=tf.tanh, name="h3")
Fhat_ij = tf.layers.dense(h3, units=3, activation=None, name="Fhat_ij")
Fhat_i = tf.reduce_sum(Fhat_ij, axis=1)

loss = tf.reduce_sum(tf.square(Fhat_i - F_i))
training_op = tf.train.AdamOptimizer(learning_rate=1E-2).minimize(loss)

# SESSION
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for epoch in t:
    index = np.random.choice(range(NI), BATCH_SIZE, replace=False)
    sess.run(training_op, 
             feed_dict = {
                 X_ij: _Xscaled_ij[index],
                 F_i: _Fscaled_i[index]
             })
    if epoch % 100 == 1:
        _Fhat_i = sess.run(Fhat_i, 
                         feed_dict = {
                             X_ij: _Xscaled_ij
                         })
        r2 = sklearn.metrics.r2_score(_Fscaled_i[:,0].reshape(-1), _Fhat_i[:,0].reshape(-1))
        r2s.append(r2)
        t.set_description(str(r2))
        ax.plot(r2s)
        fig.canvas.draw()
        fig.canvas.show()
```

# $\sum^{1000}\text{Coulomb}$, Grad | Nvidia P100

```python
NI = 1000
NJ = 500

p = np.random.rand(2) * 0.2 + 0.4
p /= np.sum(p)
_X_ij = (np.random.rand(NI, NJ, 3) * 47 + 3) * np.random.choice([-1,1], (NI, NJ, 3), p=p)
_F_i = np.sum(_X_ij / np.linalg.norm(_X_ij, axis=-1, keepdims=True)**3, axis=1)
```

Prepare for Machine Learning.

```python
_Xscaled_ij = sklearn.preprocessing.scale(_X_ij.reshape(-1,3)).reshape(NI, NJ, 3)
_Fscaled_i = sklearn.preprocessing.scale(_F_i) * NJ
```

Train.

```python
N_EPOCHS = 5000
BATCH_SIZE = 512
r2s = []
fig, ax = plt.subplots(1,1)
t = tqdm_notebook(range(N_EPOCHS))

# GRAPH
tf.reset_default_graph()
X_ij = tf.placeholder(name="X_ij", dtype=tf.float32, shape=(None, None, 3))
dX = tf.zeros(shape=(tf.shape(X_ij)[0], 3))
F_i = tf.placeholder(name="F_i", dtype=tf.float32, shape=(None, 3))

h1 = tf.layers.dense(X_ij + tf.expand_dims(dX, 1), units=16, activation=tf.tanh, name="h1")
h2 = tf.layers.dense(h1, units=20, activation=tf.tanh, name="h2")
h3 = tf.layers.dense(h2, units=32, activation=tf.tanh, name="h3")
Ehat_ij = tf.layers.dense(h3, units=1, activation=None, name="Ehat_ij")
Ehat_i = tf.reduce_sum(Ehat_ij, axis=1)
Fhat_i = tf.gradients(Ehat_i, [dX])[0]

loss = tf.reduce_sum(tf.square(Fhat_i - F_i))
training_op = tf.train.AdamOptimizer(learning_rate=1E-2).minimize(loss)
# training_op = tf.train.MomentumOptimizer(learning_rate=0.1, momentum=0.99, use_nesterov=True).minimize(loss)

# SESSION
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for epoch in t:
    index = np.random.choice(range(NI), BATCH_SIZE, replace=False)
    sess.run(training_op, 
             feed_dict = {
                 X_ij: _Xscaled_ij[index],
                 F_i: _Fscaled_i[index]
             })
    if epoch % 100 == 1:
        _Fhat_i = sess.run(Fhat_i, 
                         feed_dict = {
                             X_ij: _Xscaled_ij
                         })
        r2 = sklearn.metrics.r2_score(_Fscaled_i[:,0].reshape(-1), _Fhat_i[:,0].reshape(-1))
        r2s.append(r2)
        t.set_description(str(r2))
        ax.plot(r2s)
        fig.canvas.draw()
        fig.canvas.show()
```
