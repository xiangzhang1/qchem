# Get started with Oscar

```python
oscar = Oscar.Oscar('dPhUcQ11w087ADACkVVsg5Wj9Tx8X7O3EH6LiT6gag9zfnNlbnNvdXQtb3NjYXJyEQsSBFVzZXIYgICAgOesmAoM')
graph = {
    'name': "test",
    'parameters': {
        'b1': {
            'min': -1,
            'max': 2
        },
        'b2': {
            'min': -2,
            'max': 2
        }
    }
}
for _ in trange(50):
    sess = oscar.suggest(graph)
    b1, b2 = sess['b1'], sess['b2']
    loss = (1 - b1)**2 + 100 * (b2 - b1**2)**2
    oscar.update(sess, {'loss': loss})
```
