This notebook (attempts to) analyze a trained NN. To my dismay, the trained NN was learning something else it shouldn't.'

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import time
from tqdm import tnrange, tqdm_notebook
from sympy import Matrix
```

```python
!trash logs/*
raw = []
sess.close()
filewriter.close()
tf.reset_default_graph()

# select the model
n_instances = 1000
n_timesteps = 200
n_features = 1
n_labels = 1
hidden_RNNlayer_shape = 5

n_epochs = 1000
minibatch_size = None # batch gradient descent
learning_rate = 0.01

# get the data
_X = np.random.rand(n_instances, n_timesteps, n_features) * 2 - 1
_y = np.sum(_X, axis = 1)

# graph
with tf.variable_scope("initial"):
    X = tf.placeholder(name="X", shape=[None, n_timesteps, n_features], dtype=tf.float32)
    y = tf.placeholder(name="y", shape=[None, n_labels], dtype=tf.float32)
    h = tf.zeros(shape=[tf.shape(X)[0], hidden_RNNlayer_shape], name="h")
    n = tf.constant(0, name="n")    

with tf.variable_scope("GRU"):
    def GRU(n, h, X=X):
        Xn = X[:, n, :] # strided_slice
        hXn = tf.concat([h, Xn], axis=1, name="hXn")
        r = tf.layers.dense(hXn, units=hidden_RNNlayer_shape, activation=tf.sigmoid, name="rGate")
        z = tf.layers.dense(hXn, units=hidden_RNNlayer_shape, activation=tf.sigmoid, name="zGate")
        rhXn = tf.concat([tf.multiply(h, r), Xn], axis=1, name="rhXn")
        c = tf.layers.dense(rhXn, units=hidden_RNNlayer_shape, activation=tf.tanh, name="c")
        hnew = tf.multiply(h, z) + tf.multiply(c, 1-z)
        return [n + 1, hnew]
    n, hout = tf.while_loop(cond=lambda n,h: n<n_timesteps, body=GRU, loop_vars=[n,h])

with tf.variable_scope("final"):
    yhat = tf.layers.dense(hout, units=n_labels, activation=None, name="yhat") # 输出转换层，属于典型技术

with tf.variable_scope("training"):
    loss = tf.reduce_mean(tf.square(yhat - y), keepdims=False, name="loss")
    train_op = tf.train.AdamOptimizer(learning_rate).minimize(loss)
    summary_op = tf.summary.scalar("loss", loss)

# session
sess = tf.Session()
sess.run(tf.global_variables_initializer())
filewriter = tf.summary.FileWriter('logs/tf.teardown_run.size%s.'%n_timesteps + time.strftime("%Y%m%d%H%M%s"),
                                  tf.get_default_graph())
for epoch in tnrange(n_epochs, leave=False):
    sess.run(train_op, feed_dict={X: _X, y: _y})
    if epoch % 10 == 0:
        filewriter.add_summary(sess.run(summary_op, feed_dict={X: _X, y: _y}), epoch)
        raw.append((n_timesteps, epoch, sess.run(loss, feed_dict={X: _X, y: _y})))
```

```python
with tf.variable_scope("initial", reuse=False):
    X = tf.placeholder(name="X", shape=[1, n_features], dtype=tf.float32)
    h = tf.zeros(shape=[1, hidden_RNNlayer_shape], name="h")
    
with tf.variable_scope("GRU", reuse=True):
    Xn = X
    hXn = tf.concat([h, Xn], axis=1, name="hXn")
    r = tf.layers.dense(hXn, units=hidden_RNNlayer_shape, activation=tf.sigmoid, name="rGate", reuse=True)
    z = tf.layers.dense(hXn, units=hidden_RNNlayer_shape, activation=tf.sigmoid, name="zGate", reuse=True)
    rhXn = tf.concat([tf.multiply(h, r), Xn], axis=1, name="rhXn")
    c = tf.layers.dense(rhXn, units=hidden_RNNlayer_shape, activation=tf.tanh, name="c")
    hnew = tf.multiply(h, z) + tf.multiply(c, 1-z)

with tf.variable_scope("final", reuse=True):
    yhat = tf.layers.dense(hnew, units=n_labels, activation=None, name="yhat", reuse=True)
```
