```python
import numpy as np

# plotly
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
N = 100
random_x = np.linspace(0, 1, N)
random_y0 = np.random.randn(N)
random_y1 = np.random.randn(N)
text = ['epoch 0'] * N

# Create traces
trace0 = go.Scatter(
    x = random_x,
    y = random_y0,
    mode = 'markers',
    name = 'random_y0 $\\mu$',
    marker = dict(size=5),
    hoverinfo = 'text',
    text = text
)
trace1 = go.Scatter(
    x = random_x,
    y = random_y1,
    mode = 'markers',
    name = 'random_y1 $\\xi$',
    marker = dict(size=5),
    text = text
)
data = [trace0, trace1]
layout = go.Layout(
    title = 'test',
    hovermode = 'closest'
)
fig = go.Figure(data=data, layout=layout)
py.iplot(data)
```
